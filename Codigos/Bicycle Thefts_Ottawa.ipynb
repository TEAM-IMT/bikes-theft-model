###**Visualisation**

In general, the status has three main types: stolen, recovered and unknown. The following graph shows the proportion of each status with respect to the general data. 97% of the bicycles correspond to stolen items

```
fig=px.pie(data_frame=data_bikes,names='Bicycle_Status',title='Theft status')
fig.show()
```

***What is the trend of annual thefts ?***

In the following graph, the number of reported thefts per year is presented, it is observed that between 2014 and 2018 thefts were increasing, however a slight decrease is observed between 2018 and 2019

```
data_bikes['Occurrence_Date']=pd.to_datetime(data_bikes['Incident_Date'])
data_bikes['Occurrence_Year']=data_bikes['Occurrence_Date'].dt.year
data_bikes['Occurrence_Month']=data_bikes['Occurrence_Date'].dt.month
data_bikes['Occurrence_Day']=data_bikes['Occurrence_Date'].dt.day
data_bikes.head(5)

```

```
Dias_aux_robo_year=data_bikes.groupby(['Occurrence_Year','Bicycle_Status']).size().reset_index().rename(columns={0:'Count'})
#display(Dias_aux_robo_year)

fig = px.line(Dias_aux_robo_year, x='Occurrence_Year', y='Count',title='Vol de vélos à Ottawa', color='Bicycle_Status')
fig.show()
```

***What day do the most thefts occur ?***

According to the figure, it is observed that there is not a great difference between the number of thefts reported per day among the different days of the week, however it is observed that the greatest amount of thefts occur on Friday

```
data_bikes['Occurrence_Date']=pd.to_datetime(data_bikes['Occurrence_Date']) #Transforma la columna Formato de fecha
data_bikes['Week_day']=data_bikes['Occurrence_Date'].dt.day_name() #Agregar el nombre

day_type = pd.api.types.CategoricalDtype(categories=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"], ordered=True)
data_bikes["Week_day"] = data_bikes["Week_day"].astype(day_type)

Dias_aux_week=data_bikes[data_bikes['Bicycle_Status']=='STOLEN'].groupby(['Week_day']).size().reset_index().rename(columns={0:'Count'})
fig = px.bar(Dias_aux_week, x='Week_day', y='Count',title='Bicycle theft in Ottawa per day')
fig.show()
```

***At what time do most thefts occur?***

According to the following histogram, it is observed that the greatest amount of thefts occur between 12h and 18h

```
data_bikes['Occurrence_Time'] = pd.to_datetime(data_bikes['Occurrence_Date'].dt.time,format= '%H:%M:%S' ).dt.time
#display(data_bikes.head(5))
data_bikes['hour'] = pd.to_datetime(data_bikes['Occurrence_Time'],format= '%H:%M:%S' ).dt.hour



Dias_aux_time=data_bikes[data_bikes['Bicycle_Status']=='STOLEN'].groupby(['hour']).size().reset_index().rename(columns={0:'Count'})
fig = px.bar(Dias_aux_time, x='hour', y='Count',title='Bicycle theft in Toronto per hour')
fig.show()
```

***What are the most stolen bicycles?***

According to the information in the following table, the most stolen bicycles are those between 0 - 1000 Canadian dollars

```
bins = [1, 1000, 2000, 5000, 10000, 120000]
aux=data_bikes[data_bikes['Bicycle_Status']=='STOLEN']
aux_2=aux['Bicycle_Value'].value_counts(bins=bins, sort=False) 
aux_3=pd.DataFrame(aux_2).reset_index().rename(columns={'index':'Range','Bicycle_Value':'Total'})
aux_3
```
