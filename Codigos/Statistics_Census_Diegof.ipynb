```
def Create_DensityFrameDay(yearframe):

  #Extraction of just stolen bikes status

  #Year_aux=_Joint[_Joint['Status']=='STOLEN'].groupby(['Occurrence_Year','GeoUID']).size().reset_index().rename(columns={0:'Count'}) 
  #Aux_Geo_ot = pd.merge(Year_aux,_pop[['GeoUID','PopulationCT']], on='GeoUID', how='left')
  Theft_year_day=yearframe.groupby(['Occurrence_Day']).sum().reset_index().rename(columns={'Count':'SumThBikes'}) #Sum total per year dataframe 
  LandArea = yearframe['Area(sqKm)'].sum()
  bikes = yearframe['BicycleT'].sum()
  Population = yearframe['PopulationCT'].sum()
  

  #####NEW VARIABLES OF DENSITY ANALYSIS #########
  #Theft_year_day['BicycleT']=bikes
  Theft_year_day['BikesThDensity']=Theft_year_day['SumThBikes']/LandArea #Theft bikes density
  Theft_year_day['BikesDensity']=bikes/ LandArea#Bikes density
  Theft_year_day['PopulationDensity'] = Population/LandArea #Population Density
  #Theft_year_day['LandArea']=LandArea_
  Theft_year_day['%Theft/sqKm'] = (Theft_year_day['BikesThDensity']/Theft_year_day['BikesDensity'])*100 #Percentage of stoles bikes by sq Km
  Theft_year_day['%PopulationBike'] =(Theft_year_day['BikesDensity']/Theft_year_day['PopulationDensity'])*100 #Percentage of bikes by sq Km
  
  return Theft_year_day
```

```
Stolen_bikes_df_H2014 = CityJoint_2014[CityJoint_2014['Status']=='STOLEN'].groupby(['Occurrence_Time','PopulationCT', 'BicycleT', 'Area(sqKm)']).size().reset_index().rename(columns={0:'Count'})
Stolen_bikes_df_H2015 = CityJoint_2015[CityJoint_2015['Status']=='STOLEN'].groupby(['Occurrence_Time','PopulationCT', 'BicycleT', 'Area(sqKm)']).size().reset_index().rename(columns={0:'Count'})
Stolen_bikes_df_H2016 = CityJoint_2016[CityJoint_2016['Status']=='STOLEN'].groupby(['Occurrence_Time','PopulationCT', 'BicycleT', 'Area(sqKm)']).size().reset_index().rename(columns={0:'Count'})
Stolen_bikes_df_H2017 = CityJoint_2017[CityJoint_2017['Status']=='STOLEN'].groupby(['Occurrence_Time','PopulationCT', 'BicycleT', 'Area(sqKm)']).size().reset_index().rename(columns={0:'Count'})
Stolen_bikes_df_H2018 = CityJoint_2018[CityJoint_2018['Status']=='STOLEN'].groupby(['Occurrence_Time','PopulationCT', 'BicycleT', 'Area(sqKm)']).size().reset_index().rename(columns={0:'Count'})
Stolen_bikes_df_H2019 = CityJoint_2019[CityJoint_2019['Status']=='STOLEN'].groupby(['Occurrence_Time','PopulationCT', 'BicycleT', 'Area(sqKm)']).size().reset_index().rename(columns={0:'Count'})
```

```
def Create_DensityFrameHour(yearframe):

  #Extraction of just stolen bikes status

  yearframe['Occurrence_Time'] = pd.to_datetime(yearframe['Occurrence_Time'])
  yearframe['Hour'] = yearframe['Occurrence_Time'].dt.hour 
  Theft_year_hour=yearframe.groupby(['Hour']).sum().reset_index().rename(columns={'Count':'SumThBikes'}) #Sum total per year dataframe 
  LandArea = yearframe['Area(sqKm)'].sum()
  bikes = yearframe['BicycleT'].sum()
  Population = yearframe['PopulationCT'].sum()

  #####NEW VARIABLES OF DENSITY ANALYSIS #########
  Theft_year_hour['BikesThDensity']=Theft_year_hour['SumThBikes']/LandArea #Theft bikes density
  Theft_year_hour['BikesDensity']=bikes/ LandArea#Bikes density
  Theft_year_hour['PopulationDensity'] = Population/LandArea #Population Density
  Theft_year_hour['%Theft/sqKm'] = (Theft_year_hour['BikesThDensity']/Theft_year_hour['BikesDensity'])*100 #Percentage of stoles bikes by sq Km
  Theft_year_hour['%PopulationBike'] =(Theft_year_hour['BikesDensity']/Theft_year_hour['PopulationDensity'])*100 #Percentage of bikes by sq Km
  
  return Theft_year_hour
```

```
DensityStolenBikesHour2014=Create_DensityFrameHour(Stolen_bikes_df_H2014)
DensityStolenBikesHour2015=Create_DensityFrameHour(Stolen_bikes_df_H2015)
DensityStolenBikesHour2016=Create_DensityFrameHour(Stolen_bikes_df_H2016)
DensityStolenBikesHour2017=Create_DensityFrameHour(Stolen_bikes_df_H2017)
DensityStolenBikesHour2018=Create_DensityFrameHour(Stolen_bikes_df_H2018)
DensityStolenBikesHour2019=Create_DensityFrameHour(Stolen_bikes_df_H2019)
```

```
def Create_DensityFrameJohan(yearframe):

  #Extraction of just stolen bikes status

  Year_aux = yearframe
  Year_aux['Occurrence_Time'] = pd.to_datetime(yearframe['Occurrence_Time'])
  Year_aux['Hour'] = yearframe['Occurrence_Time'].dt.hour 
  Year_aux=yearframe[yearframe['Status']=='STOLEN'].groupby(['City','GeoUID','Hour','PopulationCT', 'BicycleT', 'Area(sqKm)']).size().reset_index().rename(columns={0:'SumThBikes'})
  #Year_aux=yearframe.groupby(['Hour']).reset_index().rename(columns={'Count':'SumThBikes'}) #Sum total per year dataframe 
  LandArea = Year_aux['Area(sqKm)']
  bikes = Year_aux['BicycleT']
  Population = Year_aux['PopulationCT']
  

  #####NEW VARIABLES OF DENSITY ANALYSIS #########
  #Theft_year_day['BicycleT']=bikes
  Year_aux['BikesThDensity']=Year_aux['SumThBikes']/Year_aux['Area(sqKm)'] #Theft bikes density
  Year_aux['BikesDensity']= Year_aux['BicycleT']/ Year_aux['Area(sqKm)']#Bikes density
  Year_aux['PopulationDensity'] =Year_aux['PopulationCT']/Year_aux['Area(sqKm)']#Population Density
  Year_aux['%Theft/sqKm'] = (Year_aux['BikesThDensity']/Year_aux['BikesDensity']) #Percentage of stoles bikes by sq Km
  Year_aux['%PopulationBike'] =(Year_aux['BikesDensity']/Year_aux['PopulationDensity']) #Percentage of bikes by sq Km
  
  return Year_aux
```

```

Stolen_bikes_johan_2014 = Create_DensityFrameJohan(CityJoint_2014)
Stolen_bikes_johan_2015 = Create_DensityFrameJohan(CityJoint_2015)
Stolen_bikes_johan_2016 = Create_DensityFrameJohan(CityJoint_2016)
Stolen_bikes_johan_2017 = Create_DensityFrameJohan(CityJoint_2017)
Stolen_bikes_johan_2018 = Create_DensityFrameJohan(CityJoint_2018)
Stolen_bikes_johan_2019 = Create_DensityFrameJohan(CityJoint_2019)
```
